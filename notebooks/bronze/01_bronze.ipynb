```python
# Widget for environment-aware catalog targeting
dbutils.widgets.text("target_catalog", "tpch_dev", "Target Catalog")
target_catalog = dbutils.widgets.get("target_catalog")

print(f"Writing bronze tables to catalog: {target_catalog}")
```

```python
source_tables = ["orders", "lineitem", "customer", "nation", "region", "supplier", "part"]
```

```python
for table in source_tables:
    df = spark.table(f"samples.tpch.{table}")
    
    target_table = f"{target_catalog}.bronze.{table}"
    
    df.write.format("delta").mode("overwrite").saveAsTable(target_table)
    
    print(f"Loaded {table}: {df.count()} rows -> {target_table}")
```
